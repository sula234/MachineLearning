### **Step 1**

```python
def vector_to_func(vec, x):
  return ((vec[1]*x) / vec[2]) + (vec[0] / vec[2])


# Generate data points
def generate_data(num_points):
  data = np.random.uniform(0, 10, (num_points, 2))
  return data
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Define the target function
f = np.random.uniform(1, 6, 3)

# Plot the target function f(x)
x_target = np.linspace(0, 11, 100)
plt.plot(x_target, vector_to_func(f, x_target), 'g', label=f'Target function: {round(f[2], 2)} y= {round(f[1], 2)}x + ({round(f[0], 2)})')

# Generate 20 data points
data = generate_data(20)
x = data[:, 0]
y = data[:, 1]

# divide points to 2 classes depended on target function
labels = [1 if y[i] >= vector_to_func(f, x[i]) else -1 for i in range(0,20)]

# separate points into 2 clases
class1_x = x[y >= vector_to_func(f, x)]
class1_y = y[y >= vector_to_func(f, x)]
class_minus1_x = x[y < vector_to_func(f, x)]
class_minus1_y = y[y < vector_to_func(f, x)]

# Plot the data points
plt.scatter(class1_x, class1_y, c='r', marker='o', label='+1 class')
plt.scatter(class_minus1_x, class_minus1_y, c='b', marker='x', label='-1 class')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper left')

# Show the plot
plt.grid(True)
plt.title('Generated Data Set and Target Function')
plt.show()


```

### **Step 2**

Define PLA Algorithm

```python
# perceptron learning algorithm
def PLA(data, labels):
    # initialize vector w by the first input data
    w = np.array([0, data[0][0], data[0][1]])
    update = True
    epochs = 0

    while update:
        epochs += 1
        update = False

        for i in range(len(data)):
            x = np.array([1, data[i][0], data[i][1]])


            # if hypothesis misclassified
            if w.dot(x) < 0 and labels[i] == 1:
                # update w
                w = w + x
                update = True
                break

            if w.dot(x) >= 0 and labels[i] == -1:
                # update w
                w = w - x
                update = True
                break

    return w, epochs




```

Get predicted function

```python
w, updates = PLA(data, labels)
```

Draw a graph along with the predicted function

```python
# Plot the target function f(x)
hyp = np.array([-1* w[0], -1 * w[1], -1 * w[2]])

plt.plot(x_target, vector_to_func(f, x_target), 'g', label=f'Target function: {round(f[2], 2)} y= {round(f[1], 2)}x +{round(f[0], 2)}')
plt.plot(x_target, vector_to_func(hyp, x_target), 'y', label=f'Predicted, updates = {updates}')


# Plot the data points
plt.scatter(class1_x, class1_y, c='r', marker='o', label='+1 class')
plt.scatter(class_minus1_x, class_minus1_y, c='b', marker='x', label='-1 class')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper left')

# Show the plot
plt.grid(True)
plt.title('Generated Data Set and Functions')
plt.show()
```
